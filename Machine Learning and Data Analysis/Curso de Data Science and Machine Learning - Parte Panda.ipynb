# GroupBy

```python
dcn = {'EMPRESAS': ['GOG','GOG','GOG','MSF','MSF','MSF','FB','FB'], 'PERSON': ['SAM','SAM','TAM','VAN','TAM','CARL',
    'TAM','SARA'],'SALES': [100,200,300,400,500,200,300,200]}
Frame = pd.DataFrame(dcn)
Frame
```

```python
bycomp = Frame.groupby('EMPRESAS')
bycomp.mean()
```

```python
bycomp.sum()
```

```python
byperson = Frame.groupby('PERSON').sum()
byperson
```

```python
bycomp.describe()
```
